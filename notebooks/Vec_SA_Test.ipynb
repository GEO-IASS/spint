```python
num = np.arange(0,50).reshape((-1,1))
OX = np.random.randint(450,500, 50).reshape((-1,1))
OY = np.random.randint(450,500, 50).reshape((-1,1))
DX = np.random.randint(0,500, 50).reshape((-1,1))
DY = np.random.randint(0,500, 50).reshape((-1,1))

vecs = np.hstack([num, OX, OY, DX, DY])
origins = vecs[:, 1:3]
dests = vecs[:, 3:5]
wo = DistanceBand(origins, threshold=9999, alpha=-1.5, binary=False)
wd = DistanceBand(dests, threshold=9999, alpha=-1.5, binary=False)


vmo = VecMoran(vecs, wo, focus='origin', rand='A')
print vmo.I
print vmo.p_z_sim
fig = plt.hist(vmo.sim, bins=25)
plt.show()

vmo = VecMoran(vecs, wo, focus='origin', rand='B')
print vmo.I
print vmo.p_z_sim
fig = plt.hist(vmo.sim, bins=25)
plt.show()


vmd = VecMoran(vecs, wd, focus='destination', rand='A')
print vmd.I
print vmd.p_z_sim
fig = plt.hist(vmd.sim, bins=25)
plt.show()

vmd = VecMoran(vecs, wd, focus='destination', rand='B')
print vmd.I
print vmd.p_z_sim
fig = plt.hist(vmd.sim, bins=25)
plt.show()

```
